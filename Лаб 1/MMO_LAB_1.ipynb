Получение датасета

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

FILE_NAME = 'movies_youtube_sentiments.csv'
df = pd.read_csv(FILE_NAME, on_bad_lines='skip', sep=',', comment='#', index_col=False)
```

График типа pie для зависимости количество фильмов по странам

```python
size_of_groups = df['country'].value_counts()
plt.pie(x=size_of_groups, labels=size_of_groups.index)
plt.show()
```

График типа Bubble plot зависимости дохода от бюджета и одобряемости

```python
data = df[['gross', 'budget', 'rating']].dropna()
data.set_index('gross', inplace=True)

sns.scatterplot(data=data, x="budget", y="rating", size="gross", legend=False, sizes=(20, 2000))
plt.show()
```

График типа Histogram зависимости количество фильмов по годам Histogram

```python
sns.displot(data=df["year"], kde=True)
plt.show()
```

График типа boxplot зависимости количества фильмов для каждого из жанров в каждом годе

```python
sns.boxplot(x=df["year"], y=df["genre"])
plt.show()
```

График типа Violin Plot зависимости дохода от жанра

```python
sns.violinplot(x=df['genre'], y=df['gross'])
plt.show()
```
